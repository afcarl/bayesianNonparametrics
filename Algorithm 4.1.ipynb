```julia
mu = 0.15
sigma = 0.5
a = mu + sigma*rand(100, 2);
```

```julia
mu = -mu
sigma = 0.4
b = mu + sigma*rand(100, 2);
```

```julia
scatter(a[:, 1], a[:, 2], label="class 1")
scatter!(b[:, 1], b[:, 2], label="class 2")
```

```julia
K = 2
data = [a; b]
num_sample, dim = size(data)
```

```julia
z = rand(1:K, num_sample);
```

```julia
function get_mu(K, data, z)
    mu_list = zeros(K, dim)

    for c in 1:K
        indices = find(z .== c)
        mu_list[c, :] = mean(data[indices, :], 1)
    end
    return mu_list
end
```

```julia
mu_list = get_mu(K, data, z)
for _ in 1:200
    for i in 1:num_sample
        x = data[i, :]
        z[i] = indmin([vecdot(x - mu_list[c, :], x - mu_list[c, :]) for c in 1:K])# indmax
    end
    mu_list = get_mu(K, data, z) 
end
```

```julia
c = 1
indices = find(z .== c)
scatter(data[indices, 1], data[indices, 2], label="z=1")

c = 2
indices = find(z .== c)
scatter!(data[indices, 1], data[indices, 2], label="z=2")
```
